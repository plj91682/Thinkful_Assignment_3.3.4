```python
# 4:Evaluate whether the descriptive statistics provided useful information about the variable. 
# Can you identify any common characteristics of the distributions that could be usefully 
# described using the mean and/or standard deviation, versus the ones that could not?
#
# Answer: The standard deviation plots help us see if the majority of the samples are within the mean.
#         Values outside of the std-deviations can highlight outlier values in the distribution.
# 
#         Since Bernoilli distributions are binary the Mean is not very meaningful.
#


import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.special as sps
%matplotlib inline

sample_size = 100
```

```python
# Generate a bernoulli distribution with p =0.5.
bernoulli_dist = np.random.binomial(1, .5, sample_size)

#Plot a histogram.
plt.hist(bernoulli)

############### plot mean & stddevs
# Add a vertical line at the mean.
plt.axvline(bernoulli_dist.mean(), color='b', linestyle='solid', linewidth=2)

# Add a vertical line at one standard deviation above the mean.
plt.axvline(bernoulli_dist.mean() + bernoulli_dist.std(), color='b', linestyle='dashed', linewidth=2)

# Add a vertical line at one standard deviation below the mean.
plt.axvline(bernoulli_dist.mean()-bernoulli_dist.std(), color='b', linestyle='dashed', linewidth=2) 


# Print the histogram
plt.show()
```

```python
# Generate a gamma distribution with shape, scale = 2.0
shape = 2.0
scale = 2.0
gamma_dist = np.random.gamma(shape,scale, sample_size)


count, bins, ignored = plt.hist(gamma_dist, 50, normed=True)
y = bins**(shape-1)*(np.exp(-bins/scale) / (sps.gamma(shape)*scale**shape))
plt.plot(bins, y, linewidth=2, color='r')


############### plot mean & stddevs
# Add a vertical line at the mean.
plt.axvline(gamma_dist.mean(), color='b', linestyle='solid', linewidth=2)

# Add a vertical line at one standard deviation above the mean.
plt.axvline(gamma_dist.mean() + gamma_dist.std(), color='b', linestyle='dashed', linewidth=2)

# Add a vertical line at one standard deviation below the mean.
plt.axvline(gamma_dist.mean()-gamma_dist.std(), color='b', linestyle='dashed', linewidth=2) 


plt.show()
```

```python
# Generate a poisson dist expectation-interval = 5
interval = 5
poisson_dist = np.random.poisson(interval, sample_size)

############### plot mean & stddevs
# Add a vertical line at the mean.
plt.axvline(poisson_dist.mean(), color='b', linestyle='solid', linewidth=2)

# Add a vertical line at one standard deviation above the mean.
plt.axvline(poisson_dist.mean() + poisson_dist.std(), color='b', linestyle='dashed', linewidth=2)

# Add a vertical line at one standard deviation below the mean.
plt.axvline(poisson_dist.mean()-poisson_dist.std(), color='b', linestyle='dashed', linewidth=2) 


count, bins, ignored = plt.hist(poisson_dist, 14, normed=True)
plt.show()
```

```python
# Generate a normal dist with mean, std-dev = 0, 0.1
mean_value, std_dev_value = 0, 0.1 # mean and standard deviation
normal_dist = np.random.normal(mean_value, std_dev_value, sample_size)

############### plot mean & stddevs
# Add a vertical line at the mean.
plt.axvline(normal_dist.mean(), color='b', linestyle='solid', linewidth=2)

# Add a vertical line at one standard deviation above the mean.
plt.axvline(normal_dist.mean() + normal_dist.std(), color='b', linestyle='dashed', linewidth=2)

# Add a vertical line at one standard deviation below the mean.
plt.axvline(normal_dist.mean()-normal_dist.std(), color='b', linestyle='dashed', linewidth=2) 

count, bins, ignored = plt.hist(normal_dist, 30, normed=True)
plt.plot(bins, 1/(std_dev_value * np.sqrt(2 * np.pi)) * np.exp( - (bins - mean_value)**2 / (2 * std_dev_value**2) ), linewidth=2, color='r')
plt.show()
```

```python
# Generate a power dist with shape = 5
shape = 5.
power_dist = np.random.power(shape, sample_size)

############### plot mean & stddevs
# Add a vertical line at the mean.
plt.axvline(power_dist.mean(), color='b', linestyle='solid', linewidth=2)

# Add a vertical line at one standard deviation above the mean.
plt.axvline(power_dist.mean() + power_dist.std(), color='b', linestyle='dashed', linewidth=2)

# Add a vertical line at one standard deviation below the mean.
plt.axvline(power_dist.mean()-power_dist.std(), color='b', linestyle='dashed', linewidth=2) 


count, bins, ignored = plt.hist(power_dist, bins=30)
x = np.linspace(0, 1, 100)
y = shape*x**(shape-1.)
normed_y = sample_size*np.diff(bins)[0]*y
plt.plot(x, normed_y)
plt.show()
```

```python
# Generate a power dist with mean, dispersion = 0.0, 4.0
mean, dispersion = 0.0, 4.0 
dispersion_dist = np.random.vonmises(mean, dispersion, sample_size)

############### plot mean & stddevs
# Add a vertical line at the mean.
plt.axvline(dispersion_dist.mean(), color='b', linestyle='solid', linewidth=2)

# Add a vertical line at one standard deviation above the mean.
plt.axvline(dispersion_dist.mean() + dispersion_dist.std(), color='b', linestyle='dashed', linewidth=2)

# Add a vertical line at one standard deviation below the mean.
plt.axvline(dispersion_dist.mean()-dispersion_dist.std(), color='b', linestyle='dashed', linewidth=2) 


from scipy.special import i0
plt.hist(dispersion_dist, 50, normed=True)
x = np.linspace(-np.pi, np.pi, num=51)
y = np.exp(dispersion*np.cos(x-mean))/(2*np.pi*i0(dispersion))
plt.plot(x, y, linewidth=2, color='r')
plt.show()
```

```python
#### Additionally, create 2 normal dist, add them together to create a 3rd dist.  
#### Graph the untion of the 1st 2 dist and plot the mean & stddevs
var1 = np.random.normal(5,.5, 100)
var2 = np.random.normal(10,1, 100)
var3 = var1+var2

mean = np.mean(var3)
sd = np.std(var3)

plt.hist(var3)
plt.axvline(x=mean,linestyle='dashed',color='black')
plt.axvline(x=mean+sd,linestyle='dashed', color='red')
plt.axvline(x=mean-sd,linestyle='dashed', color='red')
plt.show()
```
